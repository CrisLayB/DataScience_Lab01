#### 8. ¿Es conveniente hacer un análisis de componentes principales?

```python
numeric_columns = ['Age', 'Number of sexual partners', 'First sexual intercourse', 'Num of pregnancies', 
                   'Smokes (years)', 'Smokes (packs/year)', 'Hormonal Contraceptives (years)',
                   'IUD (years)', 'STDs (number)', 'STDs: Number of diagnosis']
data_numeric = data_frame[numeric_columns]
```

#### Test de Efericidad de Bartlett ####
H0: La matriz de correlación es igual a la matriz identidad <br>
HA: La matriz de correlación No es igual a la matriz identidad

```python
chi_cuadrado, p_valor = calculate_bartlett_sphericity(data_numeric)
chi_cuadrado, p_valor
```

El p-valor es menor que 0.05, por lo que se rechaza la hipótesis nula. Hay correlaciones entre las variables.

#### Índice KMO

```python
kmo_variables,kmo_modelo = calculate_kmo(data_numeric)
kmo_modelo
```

El test escasamente supera el umbral mínimo para un PCA y aún así se considerarían pobres los resultados. No se recomienda hacer PCA

#### PCA

```python
#Escalado de datos y extracción del PCA
pca_pipe = make_pipeline(StandardScaler(),PCA()) 
pca_pipe.fit(data_numeric)

#Extraccion de los componentes del pipeline
normalización = pca_pipe.named_steps['standardscaler'] 
modelo_pca = pca_pipe.named_steps['pca']

#Traslado al DF
PCA_df = pd.DataFrame(
    data = modelo_pca.components_,
    columns = data_numeric.columns,
    index = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10'])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
componentes = modelo_pca.components_
plt.imshow(componentes.T)
plt.yticks(range(len(data_numeric.columns)), data_numeric.columns)
plt.xticks(range(len(data_numeric.columns)), np.arange(modelo_pca.n_components_)+1)
plt.grid(False)
plt.colorbar();
```

```python
print(modelo_pca.explained_variance_ratio_)
```

Los resultados del PCA muestran que los primeros 8 componenetes recaban mas del 90% de la información, sin embargo al tener únicamente 10 componentes la reducción a 8 no es muy significativa por lo que confirma que hacer una PCA no resulta recomendable

#### 9. Reglas de asociación interesantes del dataset (Discretizar las variables numéricas)

Se discretizan las variables numéricas

```python
age_bin_edges = [0,17,20,25,30,40,50,60,100]
data_frame['age_cat'] = pd.cut(data_frame['Age'], 
                                    bins=age_bin_edges, labels=['Menores', '20a', '25a', '30a', '40a', '50a', 'Mayores', 'Ancianas'])
num_sex_partners_edges = [1,2,3,5,10,50]
data_frame['partners_cat'] = pd.cut(data_frame['Number of sexual partners'],
                                         bins=num_sex_partners_edges, labels=['pUna', 'pDos', 'pTres', 'p5_mas', 'p10_mas'])
first_sex_edges = [15,18,21,25,50]
data_frame['first_sex_cat'] = pd.cut(data_frame['First sexual intercourse'],
                                         bins=first_sex_edges, labels=['s_15a', 's_18a', 's_21a', 's_25_mas'])
pregnancies_edges = [1,2,3,4,10]
data_frame['pregnancies_cat'] = pd.cut(data_frame['Num of pregnancies'],
                                         bins=pregnancies_edges, labels=['emb_1B', 'emb_2B', 'emb_3B', 'emb_4_mas'])
smoke_time_edges = [1,10,20,30,50]
data_frame['smoke_time_cat'] = pd.cut(data_frame['Smokes (years)'],
                                         bins=smoke_time_edges, labels=['cigarillo_1a', 'cigarillo_10a', 'cigarillo_20a', 'cigarillo_30_mas'])
smoke_packs_edges = [1,10,20,50]
data_frame['smoke_packs_cat'] = pd.cut(data_frame['Smokes (packs/year)'],
                                         bins=smoke_packs_edges, labels=['pc_1a', 'pc_10a', 'pc_20_mas'])
hormonal_c_edges = [1,10,20,50]
data_frame['hormonal_c_cat'] = pd.cut(data_frame['Hormonal Contraceptives (years)'],
                                         bins=hormonal_c_edges, labels=['hormona_1a', 'hormona_10a', 'hormona_20_mas'])
iud_edges = [1,2,5,10,20]
data_frame['iud_cat'] = pd.cut(data_frame['IUD (years)'],
                                         bins=iud_edges, labels=['iud_1a', 'iud_2a', 'iud_5a', 'iud_10_mas'])
std_edges = [1,2,3,5]
data_frame['std_cat'] = pd.cut(data_frame['STDs (number)'],
                                         bins=std_edges, labels=['std_1', 'std_2', 'std_3_mas'])
std_d_edges = [1,2,3,5]
data_frame['std_d_cat'] = pd.cut(data_frame['STDs: Number of diagnosis'],
                                         bins=std_d_edges, labels=['diag_1', 'diag_2', 'diag_3_mas'])
```

```python
lista_col_cat = ['STDs:condylomatosis',
       'STDs:cervical condylomatosis', 'STDs:vaginal condylomatosis',
       'STDs:vulvo-perineal condylomatosis', 'STDs:syphilis',
       'STDs:pelvic inflammatory disease', 'STDs:genital herpes',
       'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV',
       'STDs:Hepatitis B', 'STDs:HPV', 'Dx:Cancer', 'Dx:CIN', 'Dx:HPV', 'Dx', 'Hinselmann', 'Schiller',
       'Citology', 'Biopsy', 'age_cat', 'partners_cat', 'first_sex_cat',
       'pregnancies_cat', 'smoke_time_cat', 'smoke_packs_cat', 'Hormonal Contraceptives',
       'hormonal_c_cat', 'iud_cat', 'std_cat', 'std_d_cat']
categoricas = data_frame[lista_col_cat]
```
